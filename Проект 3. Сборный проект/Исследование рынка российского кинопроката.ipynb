## Объединение двух датафреймов

Нам необходимо объединить два датафрейма по столбцу **puNumber**. У данного столбца разный тип в таблицах, необходимо привести к одинаковому типу данных для корректного объединения.

## Предобработка данных

Создадим несколько функций, к которым будем обращаться в дальнейшем.

```python
# создадим две функции для проверки значений в столбцах датафрейма
def checkup_count(column):
    return df[column].value_counts()

def checkup_unique(column):
    return df[column].sort_values().unique()
# создадим две функции для проверки значений в столбцах датафрейма

# создадим функцию для подсчета пропущенных значений
def checkup_na(column):
    return df[column].isna().sum()

# создадим функцию с двумя параметрами, которая будет изменять тип данных в определенном столбце
def change_type(column, kind):
    df[column] = df[column].astype(kind)
```

#### Столбцы refundable_support, nonrefundable_support, budget, financing_source

Первоначально рассмотрим ячейки без пропущенных значений столбцов **refundable_support**, **nonrefundable_support**, **budget**, **financing_source**.

```python
df.loc[df['refundable_support'].isna() == False].head(10) 
```

Изменим тип данных в столбцах **refundable_support**, **nonrefundable_support**, **budget** на целочисленный (**Int64**), а тип данных столбца **financing_source** оставим без изменений.

Обратимся к ранее созданной функции и изменим тип данных этих столбцов.

```python
change_type('refundable_support', 'Int64')
change_type('nonrefundable_support', 'Int64')
change_type('budget', 'Int64') # заменим типы данных в столбцах на Int64

df[['refundable_support', 'nonrefundable_support', 'budget']].dtypes # проверим
```

#### Столбец ratings

Столбец **ratings** с типом **object** необходимо преобразовать в **float64**. Посмотрим какое количество строк содержит знак % в столбце **ratings**.

```python
len(df[df['ratings'].str.contains('%', na=False)]) # проверим
```

Судя по КиноПоиску, рейтинг в % - рейтинг ожидания для еще не вышедших в прокат фильмов / сериалов. Соответственно рейтинг фильма в прокате - число от 0 до 10 с плавающей точкой. Не будем удалять данные **29** некорректных значений, а заменю методом `to_numeric` эти значения на **NaN**.

```python
df['ratings'] = pd.to_numeric(df['ratings'], errors='coerce') 
# применим метод 'to_nomeric' с параметром 'errors=coerce', для замены некорректных значений на 'NaN'

df['ratings'].dtypes # проверим
```

Обратимся к ранее созданной функции и посмотрим количество самых частых значений столбца **genres**.

Значения пересекаются, перечисляются через запятую. Жанры сложно категорировать.

Существует два вида жанров фильмов:

1) когда жанр состоит из одного слова, к примеру **драма**, **боевик**, **комедия** и т.д.;
2) составной - когда жанр фильма состоит из нескольких слов (жанров), при этом основным жанром является первое слово в значении.

Чтобы добавить новый столбец с информацией о основном жанром необходимо создать функцию, которая будет разделать значения запятой в список и возвращать значение с нулевым индексом.

```python
def n_genres(row):
    splitted_text = row.split(",")
    return splitted_text[0]

df['new_genres'] = df['genres'].apply(n_genres)
```

Проверим, для этого сгруппируем по новому столбцу **new_genres**.

```python
df.groupby('new_genres')['new_genres'].agg('count').sort_values(ascending=False)
```

```python
genres_to_rating = df.pivot_table(
    index = 'new_genres',
    values = 'ratings',
    aggfunc = ['count', 'mean', 'median']
)

genres_to_rating
```

### Выход фильмов в прокат ежегодно

Чтобы посмотреть какое количество фильмов выходило каждый год: построим сводную таблицу и построим график.

```python
general_release = df.pivot_table(
    index = 'year_release',
    values = 'title',
    aggfunc = 'count',
) # создадим сводную таблицу

general_release.head(10).style.background_gradient('Blues') # выведем результат на экран
```

```python
df.pivot_table(index='year_release', values='title', aggfunc='count').plot(
    grid=True, 
    figsize=(12, 8), 
    style='o-',
    title='Динамика выхода фильмов в прокат' 
)

plt.xlabel('Год выхода в прокат')
plt.ylabel('Количество')
plt.ylim(0);
```

Исследование показало, что в **2017** году был спад в количестве фильмов вышедших в прокат, но к следующему году ситуация выровнялась. Мы можем видеть, что количество фильмов, выходящих в прокат, ежегодно растет.

Выделим фильмы, с информацией о прокате в кинотеатрах.

```python
release_with_info = df.loc[df['box_office'] > 0].pivot_table(
    index = 'year_release',
    values = 'title',
    aggfunc = 'count',
) # создадим сводную таблицу

release_with_info.head(10).style.background_gradient('Blues') # выведем результат на экран
```

Для того чтобы посчитать долю фильмов с указанной информацией о прокате в кинотеатрах: соберем две сводные таблицы в одну (**union**).

```python
union = general_release.merge(release_with_info, on='year_release', how='left') 
# 'year_relase' - столбец, по значениям которого будем объединять

union.style.background_gradient('Blues') # посмотрим
```

Посчитаем долю фильмов с указанной информацией о прокате в кинотеатрах к общей, для этого создадим новый столбец **proportion**.

```python
union['proportion'] = union['title_y'] / union['title_x'] # создаем новый столбец

union.style.background_gradient('Blues') # выведем результат на экран
```

Мы видим, что доля фильмов с указанной информацией о прокате в кинотеатрах ежегодно увеличивается, но в **2018** и **2019** голах пошла на спад.

Посчитаем общую долю фильмов с указанной информацией о прокате в кинотеатрах.

```python
common = union['title_y'].sum() / union['title_x'].sum() # посчитаем общую долю

print(f'Общая доля фильмов с указанной информацией о прокате: {common:.2%}') # выведем результат
```

### Анализ влияния возрастного ограничения на сборы

Построим сводную таблицу для подсчета суммы сборов фильмов в прокате в период с 2015 по 2019 год.

```python
age_res = df.loc[df['year_release'] >= 2015].pivot_table(
    index = 'age_restriction',
    columns = 'year_release',
    values = 'box_office',
    aggfunc = 'sum',
) # построим сводную таблицу

age_res # выведем результат
```

Посчитаем сумму сборов для каждой категории возрастного ограничения в период с 2015 по 2019 года.

```python
age_res.columns = ['2015', '2016', '2017', '2018', '2019'] # переименуем названия столбцов
age_res['sum'] = age_res['2015'] + age_res['2016'] + age_res['2017'] + age_res['2018'] + age_res['2019'] 
# создаим новый столбец с суммой значений по каждой категории

age_res # посмотрим результат
```

Построим столбчатые диаграммы для значений каждого года.

Создадим функцию, которая будет строить столбчатую диаграмму согласно значению **column** с нужными нам параметрами.

```python
def construct_bins(column):
    age_res.plot(
    y=column, 
    kind='bar', 
    figsize =(8, 5)
)

    plt.title('Сумма сборов к категориям возрастного ограничения в ' + column)
    plt.xlabel('Категории')
    plt.ylabel('Сумма, ₽')
    plt.grid()
    plt.show()
```

Теперь нам необходим список столбцов, для значений из которых мы хотим построить столбчатую диаграмму. Сделаем цикл, который будет проходить по значениям нашего списка и инициировать запуск функции.

```python
bins_list = ['2015', '2016', '2017', '2018', '2019', 'sum'] # создаим список, к которому будем обращаться в цикле

for i in bins_list:
    construct_bins(i) # цикл, проходящий по списку и инициирующий запуск функции
```

**Выводы:**

- больше всего денег в прокате собрали фильмы в категории **16+**;
- у каждой возрастной категории сумма сборов поднималась до **2018** года, в **2018** году и после динамика пошла на спад. Предположу, что это связано с мировым финансовым кризисом **2018** года, картин стали снимать меньше и, соответственно меньше собирать средств в прокате.
