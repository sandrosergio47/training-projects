# Описание проекта

Заказчик проекта - компания федеральный оператор сотовой связи, который предлагает клиентам два тарифных плана: «Смарт» и «Ультра». Чтобы скорректировать рекламный бюджет, коммерческий департамент хочет понять, какой тариф приносит больше денег.

**Задачи проекта:**

- провести предварительный анализ тарифов на небольшой выборке клиентов;
- проанализировать поведение клиентов и подготовить вывод - какой тариф лучше.

**Описание тарифов**

**Тариф «Смарт»**

- Ежемесячная плата: 550 рублей
- Включено 500 минут разговора, 50 сообщений и 15 Гб интернет-трафика
- Стоимость услуг сверх тарифного пакета: 
    1. минута разговора: 3 рубля (оператор связи всегда округляет вверх значения минут и  мегабайтов. Если пользователь проговорил всего 1 секунду, в тарифе засчитывается целая минута); 
    2. сообщение: 3 рубля; 
    3. 1 Гб интернет-трафика: 200 рублей.

**Тариф «Ультра»**

- Ежемесячная плата: 1950 рублей
- Включено 3000 минут разговора, 1000 сообщений и 30 Гб интернет-трафика
- Стоимость услуг сверх тарифного пакета: 
    1. минута разговора: 1 рубль; 
    2. сообщение: 1 рубль; 
    3. 1 Гб интернет-трафика: 150 рублей.

**Примечания:**

- Сотовый оператор всегда округляет секунды до минут, а мегабайты — до гигабайт. Каждый звонок округляется отдельно: даже если он длился всего 1 секунду, будет засчитан как 1 минута.
- Для веб-трафика отдельные сессии не считаются. Вместо этого общая сумма за месяц округляется в бо́льшую сторону. Если абонент использует 1025 мегабайт в этом месяце, с него возьмут плату за 2 гигабайта.

### Преобразование в формат даты и времени

Создадим функцию и преобразуем данные с датой в формат **datetime**.

```python
# создадим функцию, с двумя параметрами для изменения на формат 'datetime'
def change_date(df, column):
    time_format = '%Y-%m-%d' 
    df[column] = pd.to_datetime(df[column], format = time_format)
```

Инициируем функцию.

```python
change_date(calls, 'call_date')
change_date(internet, 'session_date')
change_date(messages, 'message_date')
change_date(users, 'churn_date')
change_date(users, 'reg_date')
```

Проверим формат данных в столбцах.

### Расчет показетелей пользователей

Посчитаем количество звонков, минут, сообщений, мегабайт интернет-траффика, которые израсходовал каждый пользователей по месяцам. 

Так как мы будем смотреть информацию по месяцам, первоначально усечем даты до месяца у первоначальных таблиц.

```python
# округлим даты до месяца
calls['call_date'] = calls['call_date'] - pd.offsets.MonthBegin(1)
messages['message_date'] = messages['message_date'] - pd.offsets.MonthBegin(1)
internet['session_date'] = internet['session_date'] - pd.offsets.MonthBegin(1)

messages.head() # проверим
```

Сгруппируем таблицы, чтобы посмотреть показатели пользователей по месяцам.

```python
calls_grouped = (calls
                 .groupby(['user_id', 'call_date'])
                 .agg({'duration':'sum', 'user_id':'count'})
                 .rename(columns={'duration':'duration_sum', 'user_id':'calls_count'})
                 .reset_index()
                ) # сгруппируем по идентификатору пользователя и дате, посчитаем сумму минут и количество звонков каждого 
# пользователя, переименуем столбцы и избежим мульти-индексов

calls_grouped.head(10) # посмотрим
```

Сформируем таблицу с информацией о сообщениях пользователей.

```python
messages_grouped = (messages
                    .groupby(['user_id', 'message_date'])
                    .agg({'id':'count'})
                    .rename(columns={'id':'messages_count'})
                    .reset_index()
                   ) # формируем таблицу

messages_grouped.head(10) # посмотрим
```

Сгруппируем данные с информацией об интернет-трафике.

```python
internet_grouped = (internet
                    .groupby(['user_id', 'session_date'])
                    .agg({'mb_used':'sum'})
                    .rename(columns={'mb_used':'mb_sum'})
                    .reset_index()
                   )

internet_grouped.head(10) # проверим
```

Объединим полученный результат в одну таблицу. Присоединим наши сформированные таблицы по идентификатору пользователя и дате.

```python
calls_grouped = calls_grouped.rename(columns={'call_date':'month'})
messages_grouped = messages_grouped.rename(columns={'message_date':'month'})
internet_grouped = internet_grouped.rename(columns={'session_date':'month'}) # переименуем столбцы с информацией о месяце

# создадим переменную df, в которой сохраним нашу собранную таблицу
df = calls_grouped.merge(messages_grouped, on=['user_id','month'], how='outer')
df = internet_grouped.merge(df, on=['user_id','month'], how='outer')

display(df.head(10)) # посмотрим

df.info()
```

Появились пропущенные значения, так как не все пользователи отправляли сообщения или использовали звонки. 

К получившемуся датафрейму добавим таблицу **users**, чтобы иметь информацию о тарифном плане и городе проживания пользователя.

```python
df = users.merge(df, on='user_id', how='left') 

df # посмотрим
```

К предыдущей таблице добавилось две строки с пустыми значениями в полях **month**, **mb_sum**, **duration_sum**, **calls_count**, **messages_count**. Удалим данные строки, это действие не исказит результаты исследований.

```python
df.loc[df['month'].isna()] # посмотрим на эти строки
```

```python
df = df.drop([833, 2349]).reset_index(drop=True) # удалим
```

У нас получилась таблица, в которой для каждого пользователя посчитаны: количество звонков, израсходованных минут, отправленных сообщений и объема использованного интернет-траффика по месяцам. Также в таблице фигурирует информация о месте жительства, тарифе, возрасте и другая информация о пользователе.

### Расчет помесячной выручки с каждого пользователя

Посчитаем какое количество трафика израсходовал пользователь свыше лимита тарифа помесячно.

Создадим дублирующий столбец, из которого будем вычитать лимиту трафика в месяц согласно тарифу.

```python
df['gb_overspending'] = df['mb_sum'] # создадим столбец

df.loc[df['tariff'] == 'smart', 'gb_overspending'] = df.loc[df['tariff'] == 'smart', 'gb_overspending'] - 15360 
# вычтем 15360 мб трафика у пользователей с тарифом 'smart'
df.loc[df['tariff'] == 'ultra', 'gb_overspending'] = df.loc[df['tariff'] == 'ultra', 'gb_overspending'] - 30720
# повторим эти действия для пользователей с тарифом 'ultra'

df.head(10) # посмотрим
```

Отрицательные значения в столбце **gb_overspending** означают что пользователь не израсходовал мегабайты интернет-трафика в этом месяце. Заменим отрицательные значения в данном столбце на ноль.

```python
df.loc[df['gb_overspending'] < 0, 'gb_overspending'] = 0 # заменим отрицательные значения на ноль
```

За дополнительную плату пользователю дается один гигабайт интернет-траффика. Переведем количество перерасходованного интернет-трафика в гигабайты и округлим в большую сторону.

```python
df['gb_overspending'] = df['gb_overspending'] / 1024 # определим количество гигабайтов
df['gb_overspending'] = df['gb_overspending'].apply(np.ceil) # округлим в большую сторону

df.head(10) # проверим
```

Далее посчитаем помесячный перерасход минут разговора и сообщений для каждого пользователя.

```python
df['duration_overspending'] = df['duration_sum'] # продублируем столбец с продолжительностью звонков

df.loc[df['tariff'] == 'smart', 'duration_overspending'] = df.loc[df['tariff'] == 'smart', 'duration_overspending'] - 500 
# вычтем 500 минут разговоров у пользователей с тарифом 'smart'
df.loc[df['tariff'] == 'ultra', 'duration_overspending'] = df.loc[df['tariff'] == 'ultra', 'duration_overspending'] - 3000
# повторим эти действия для пользователей с тарифом 'ultra'

df.head(10) # посмотрим
```

Отрицательные значения в столбце **duration_overspending** означают что пользователь не израсходовал лимит минут разговора в этом месяце. Заменим отрицательные значения в данном столбце на ноль.

```python
df.loc[df['duration_overspending'] < 0, 'duration_overspending'] = 0 # заменим отрицательные значения на ноль
```

```python
df['messages_overspending'] = df['messages_count'] # продублируем столбец с количеством сообщений

df.loc[df['tariff'] == 'smart', 'messages_overspending'] = df.loc[df['tariff'] == 'smart', 'messages_overspending'] - 50 
# вычтем 50 сообщений у пользователей с тарифом 'smart'
df.loc[df['tariff'] == 'ultra', 'messages_overspending'] = df.loc[df['tariff'] == 'ultra', 'messages_overspending'] - 1000
# повторим эти действия для пользователей с тарифом 'ultra'

df.head(10) # посмотрим
```

Отрицательные значения в столбце **messages_overspending** означают что пользователь не израсходовал лимит сообщений в этом месяце. Заменим отрицательные значения в данном столбце на ноль.

```python
df.loc[df['messages_overspending'] < 0, 'messages_overspending'] = 0 # заменим отрицательные значения на ноль
```

Заменим пропущенные значения на ноль в столбцах **gb_overspending**, **duration_overspending** и **messages_overspending**, чтобы избежать возникновения проблем при подсчете стоимость каждого пользователя помесячно.

### Расчет дисперсии и стандартного отклонения

Для подсчета дисперсии и стандартного отклонения воспользуемся методами `np.var()` и `np.std()`. При этом не забудем, что мы работаем с выборкой, поэтому мы будем считать **оцененную дисперсию** и **оцененное стандартное отклонение**, добавив параметр `ddof=1`.

```python
# посчитаем оцененные дисперсии и выведем результат на экран
variance_estimate_mb_smart = np.var(df_smart['mb_sum'], ddof=1)
print(f'Оцененная дисперсия для объёма интернет-трафика, тариф smart: {variance_estimate_mb_smart:.2f}')

variance_estimate_mb_ultra = np.var(df_ultra['mb_sum'], ddof=1)
print(f'Оцененная дисперсия для объёма интернет-трафика, тариф ultra: {variance_estimate_mb_ultra:.2f}')

variance_estimate_dur_smart = np.var(df_smart['duration_sum'], ddof=1)
print(f'Оцененная дисперсия для количества минут разговора, тариф smart: {variance_estimate_dur_smart:.2f}')

variance_estimate_dur_ultra = np.var(df_ultra['duration_sum'], ddof=1)
print(f'Оцененная дисперсия для количества минут разговора, тариф ultra: {variance_estimate_dur_ultra:.2f}')

variance_estimate_mess_smart = np.var(df_smart['messages_count'], ddof=1)
print(f'Оцененная дисперсия для количества сообщений, тариф smart: {variance_estimate_mess_smart:.2f}')

variance_estimate_mess_ultra = np.var(df_ultra['messages_count'], ddof=1)
print(f'Оцененная дисперсия для количества сообщений, тариф ultra: {variance_estimate_mess_ultra:.2f}')
```

Далее посчитаем стандартное отклонение для наших значений.

```python
# посчитаем оцененные дисперсии и выведем результат на экран
standard_dev_mb_smart = np.std(df_smart['mb_sum'], ddof=1)
print(f'Стандартное отклонение для объёма интернет-трафика, тариф smart: {standard_dev_mb_smart:.2f}')

standard_dev_mb_ultra = np.std(df_ultra['mb_sum'], ddof=1)
print(f'Стандартное отклонение для объёма интернет-трафика, тариф ultra: {standard_dev_mb_ultra:.2f}')

standard_dev_dur_smart = np.std(df_smart['duration_sum'], ddof=1)
print(f'Стандартное отклонение для количества минут разговора, тариф smart: {standard_dev_dur_smart:.2f}')

standard_dev_dur_ultra = np.std(df_ultra['duration_sum'], ddof=1)
print(f'Стандартное отклонение для количества минут разговора, тариф ultra: {standard_dev_dur_ultra:.2f}')

standard_dev_mess_smart = np.std(df_smart['messages_count'], ddof=1)
print(f'Стандартное отклонение для количества сообщений, тариф smart: {standard_dev_mess_smart:.2f}')

standard_dev_mess_ultra = np.std(df_ultra['messages_count'], ddof=1)
print(f'Стандартное отклонение для количества сообщений, тариф ultra: {standard_dev_mess_ultra:.2f}')
```

**Промежуточный вывод:**

- показатели оцененной дисперсии и стандартного отклонения выше у пользователей с тарифом **ultra**, что подтверждают графики распределения плотностей.

## Проверка гипотез

### Гипотеза 1

Сформулируем нулевую и альтернативную гипотезы.

H0: Средняя выручка пользователей тарифа Ультра равна средней выручки пользователей тарифа Смарт.

H1: Средняя выручка пользователей тарифа Ультра **не** равна средней выручки пользователей тарифа Смарт.

Альтернативная гипотеза является двусторонней.

Параметр `equal_var` укажем как **False**, т.к. есть повод усомниться в равенстве дисперсий.

Критический уровень статистической значимости укажем как **.05**.

```python
alpha = 0.05 # зададим критический уровень статистической значимости

results1 = st.ttest_ind(
    df.loc[df['tariff'] == 'smart', 'value'], 
    df.loc[df['tariff'] == 'ultra', 'value'],
    equal_var = False
) # проверим гипотезу о равенстве двух независимых совокупностей

print('p-значение:', results1.pvalue)

if results1.pvalue < alpha: # сравним получившееся p-значение с заданным уровнем статистической значимости
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Промежуточный вывод:**

- статистическое исследование показало, что средняя выручка пользователей тарифа **ultra** различается от средней выручки пользователей тарифа **smart**.

### Гипотеза 2

Сформулируем нулевую и альтернативную гипотезы.

H0: Средняя выручка пользователей из Москвы равна средней выручки пользователей из других регионов.

H1: Средняя выручка пользователей из Москвы **не** средней выручки пользователей из других регионов.

Альтернативная гипотеза является двусторонней.

Параметр `equal_var` укажем как **False**, т.к. есть повод усомниться в равенстве дисперсий.

Критический уровень статистической значимости укажем как **.05**.

```python
results2 = st.ttest_ind(
    df.loc[df['city'] == 'Москва', 'value'], 
    df.loc[df['city'] != 'Москва', 'value'],
    equal_var = False
) # проверим гипотезу о равенстве двух независимых совокупностей

print('p-значение:', results2.pvalue)

if results2.pvalue < alpha: # сравним получившееся p-значение с заданным уровнем статистической значимости
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Промежуточный вывод:**

- статистическое исследование показало, что мы не можем опровергнуть нулевую гипотезу, средняя выручка пользователей из Москвы может не отличаться от средней выручки пользователей из других регионов.
