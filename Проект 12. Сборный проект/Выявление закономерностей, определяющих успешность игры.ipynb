## Предобработка данных

### Изменение названий столбцов

Приведем названия столбцов согласно "змеиному_регистру".

```python
# уберем заглавные буквы с названий столбцов
df = df.rename(str.lower, axis='columns')

df.columns # проверим
```

### Преобразование значений в столбце user_score

Проверим значения в столбце **user_score**

```python
df['user_score'].unique() # посмотрим
```

Аббревиатура **tbd** ('To Be Determined') - будет анонсировано. Выходит, что игра с данным значением в столбце **user_score** ещё не успела получить пользовательскую оценку. Для нас это равноценно отсутствию баллов за игру, мы изменим данные значения как пропущенные.

```python
df.loc[df['user_score'] == 'tbd', 'user_score'] = np.nan # заменим на 'nan'

df.loc[df['user_score'] == 'tbd', 'user_score'].count() # проверим
```

### Обработка пропусков

Первоначально удалим строки с пропущенными значениями в столбцах **Name**, **Year_of_Release** и **Genre**. Данные строки составляют **1.63%** от всего датафрейма, удаление которых не приведет к искажению будущих исследований.

```python
df.dropna(subset=['name', 'year_of_release', 'genre'], inplace=True) # удалим строки с пропущенными значениями

pd.DataFrame(round((df[['name', 'year_of_release', 'genre']].isna().mean()*100),2)).style.background_gradient() # проверим
```

Вернемся к пропущенным значениям в столбцах **critic_score**, **user_score**, **rating** и посмотрим процент пропущенных значений.

```python
pd.DataFrame(round((df[['critic_score', 'user_score', 'rating']].isna().mean()*100),2)).style.background_gradient() # посмотрим
```

В столбцах **critic_score**, **user_score**, **rating** большой процент пропущенных значений (**51.45**, **54.62** и **40,6** соответственно). Чтобы избежать искажений в дальнейших исследованиях, оставим данные пропуски без изменений.

### Преобразование типов данных

Преобразуем типы данных в столбцах **year_of_release** и **user_score** на **int64** и **float64** соответственно.

```python
df['year_of_release'] = df['year_of_release'].astype('int64')
df['user_score'] = df['user_score'].astype('float64') # преобразуем

df[['year_of_release', 'user_score']].info() # посмотрим
```

### Поиск явных дубликатов

Посмотрим есть ли явные дубликаты в таблице.

```python
print('Дубликатов в таблице:', df.duplicated().sum()) # проверим
```

### Расчет суммарных продаж

Рассчитаем суммарные продажи во всех регионах и добавим новый столбец (**total_sales**) в таблицу.

```python
df['total_sales'] = df[['na_sales','eu_sales','jp_sales', 'other_sales']].sum(axis = 1) # просуммируем

df.head(10) # проверим
```

## Исследовательский анализ данных

### Расчет количества выпущенных игр по годам

Посмотрим какое количество игр выпускалось по годам. Для этого сгруппируем данные.

```python
df.groupby('year_of_release')['name'].agg({'count'}).T # посмотрим
```

**Промежуточные выводы:**

- для прогноза продаж на 2017 год нам **не** важны данные за весь период;
- после **2009** года произошел спад по количеству выпущенных игр, возможно это связано с переходом на новое поколение консолей.

### Анализ динамики продаж копий игр по платформам и годам

Посмотрим на платформы с наибольшими продажами. Для этого сгруппируем данные.

```python
df.groupby('platform')['total_sales'].agg('sum').sort_values(ascending=False).head(10) # посмотрим
```

Выделим топ-5 платформ, по суммарным продажам за все время, и посмотрим динамику проданных игр по годам.

```python
# выделим и сгруппируем данные платформ из топ-5
ps2 = df[df['platform']=='PS2'][['year_of_release', 'total_sales']].groupby('year_of_release').agg('sum').reset_index()
x360 = df[df['platform']=='X360'][['year_of_release', 'total_sales']].groupby('year_of_release').agg('sum').reset_index()
ps3 = df[df['platform']=='PS3'][['year_of_release', 'total_sales']].groupby('year_of_release').agg('sum').reset_index()
wii = df[df['platform']=='Wii'][['year_of_release', 'total_sales']].groupby('year_of_release').agg('sum').reset_index()
ds = df[df['platform']=='DS'][['year_of_release', 'total_sales']].groupby('year_of_release').agg('sum').reset_index()
```

```python
# построим график динамики изменений проданных копий по годам на платформах из топ-5
plt.figure(figsize=(15, 8))

plt.plot(ps2['year_of_release'], ps2['total_sales'], label='PS2')
plt.plot(x360['year_of_release'], x360['total_sales'], label='X360')
plt.plot(ps3['year_of_release'], ps3['total_sales'], label='PS3')
plt.plot(wii['year_of_release'], wii['total_sales'], label='Wii')
plt.plot(ds['year_of_release'], ds['total_sales'], label='DS')


plt.xlabel('Год')
plt.ylabel('Суммарные продажи, млн. копий')
plt.grid(which='major', axis='both')
plt.title('Динамика изменений продаж по платформам')
plt.legend()
plt.show();
```

**Промежуточные выводы:**

- платформа с самым большим количеством проданных игр является **PS2**;
- в топ-5 попали платформы, поколению которым уже пришла замена;
- жизненный цикл платформы составляет около **7-8** лет.

### Определение актуального периода

Для построения прогноза на **2017** год возьмем данные за предыдущие **3** года. Из-за того, что данные за **2016** год могут быть не полными, определим актуальный период с **2013** года включительно.

```python
actual = df.loc[df['year_of_release'] >= 2013] # выделим данные с 2013 года включительно

actual.head(5) # посмотрим
```

### Определение потенциально прибыльных платформ

Первоначально узнаем на какие платформы продолжают продаваться копии игр после **2012** года.

```python
actual['platform'].unique() # посмотрим
```

Выделим платформы из списка, которые не фигурировали в топ-5 проданных копий игр за все время.

```python
# выделим и сгруппируем данные для определенных платформ
ps4 = actual[actual['platform']=='PS4'][['year_of_release', 'total_sales']].groupby('year_of_release').agg('sum').reset_index()
ds3 = actual[actual['platform']=='3DS'][['year_of_release', 'total_sales']].groupby('year_of_release').agg('sum').reset_index()
xone = actual[actual['platform']=='XOne'][['year_of_release', 'total_sales']].groupby('year_of_release').agg('sum').reset_index()
wiiu = actual[actual['platform']=='WiiU'][['year_of_release', 'total_sales']].groupby('year_of_release').agg('sum').reset_index()
pc = actual[actual['platform']=='PC'][['year_of_release', 'total_sales']].groupby('year_of_release').agg('sum').reset_index()
psv = actual[actual['platform']=='PSV'][['year_of_release', 'total_sales']].groupby('year_of_release').agg('sum').reset_index()
```

```python
# построим график динамики изменений проданных копий по годам на выделенных платформах
plt.figure(figsize=(15, 8))

plt.plot(ps4['year_of_release'], ps4['total_sales'], label='PS4')
plt.plot(ds3['year_of_release'], ds3['total_sales'], label='3DS')
plt.plot(xone['year_of_release'], xone['total_sales'], label='XOne')
plt.plot(wiiu['year_of_release'], wiiu['total_sales'], label='WiiU')
plt.plot(pc['year_of_release'], pc['total_sales'], label='PC')
plt.plot(psv['year_of_release'], psv['total_sales'], label='PSV')

plt.xlabel('Год')
plt.ylabel('Суммарные продажи, млн. копий')
plt.grid(which='major', axis='both')
plt.title('Динамика изменений продаж по платформам')
plt.legend()
plt.show();
```

**Промежуточные выводы:**

- график показал, что динамика продаж по годам падает. Так как данные платформы являются актуальными поколениями для текущего периода, вероятно уменьшение продаж связано с переходом на цифровые версии игр;
- лидирует **PS4**, также стоит отметить платформу **XOne**. Несмотря на падающую динамику данных платформ определим их как **потенциально прибыльные**.

### Анализ диаграммы размаха продаж игр в разбивке по платформам

Построим диаграммы размаха с продажами игр в разбивке по платформам.

```python
# построим диаграммы размаха
plt.figure(figsize=(15, 8))

sns.boxplot(data = actual, x='total_sales', y='platform')

plt.grid(which='major', axis='both')

plt.xlabel('Продажи, млн. копий')
plt.ylabel('Платформа')
plt.show()
```

На графике не видны медиана и квантили. Ограничим ось для большей детализации.

```python
# построим диаграммы размаха
plt.figure(figsize=(15, 8))

sns.boxplot(data = actual, x='total_sales', y='platform')

plt.grid(which='major', axis='both')

plt.xlim([0, 2])
plt.xlabel('Продажи, млн. копий')
plt.ylabel('Платформа')
plt.show()
```

**Промежуточные выводы:**

- платформа, на которой больше всего было продано копий популярной игры оказалась **PS3** с показателем в **21.05 млн.**;
- как правило на определенную платформу игра продаются менее чем **1.75 млн.** копий;
- отмечу **X360**, **PS4**, на данных платформах чаще чем у других продавались игры более **1 млн.** копий.

### Оценка влияния оценок на суммарные продажи

Рассмотрим, как влияют на продажу внутри нескольких популярных платформ отзывы пользователей и критиков. Для этого посчитаем корреляцию для платформ **PS4** и **XOne**.

## Проверка гипотез

### Гипотеза 1

Сформулируем нулевую и альтернативную гипотезы.

H0: Средний пользовательский рейтинг платформы **XOne** равен среднему пользовательскому рейтингу платформы **PC**;

H1: Средний пользовательский рейтинг платформы **XOne** **не** равен среднему пользовательскому рейтингу платформы **PC**.

Альтернативная гипотеза является двусторонней.

Возьмем специальный метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей `scipy.stats.ttest_ind()`.

Параметр `equal_var` укажем как **False**, т.к. есть повод усомниться в равенстве дисперсий.

Параметр `nan_policy` - **omit** даст возможность обойти пропущенные значения.

Критический уровень статистической значимости укажем как **.05**.

```python
alpha = 0.05 # зададим критический уровень статистической значимости

results1 = st.ttest_ind(
    df.loc[df['platform'] == 'XOne', 'user_score'], 
    df.loc[df['platform'] == 'PC', 'user_score'],
    equal_var = False,
    nan_policy = 'omit'
) # проверим гипотезу о равенстве двух независимых совокупностей

print('p-значение:', results1.pvalue)

if results1.pvalue < alpha: # сравним получившееся p-значение с заданным уровнем статистической значимости
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Промежуточный вывод:**

- статистическое исследование показало, что средний пользовательский рейтинг платформы **XOne** различается от среднего пользовательского рейтинга платформы **PC**.

### Гипотеза 2

Сформулируем нулевую и альтернативную гипотезы.

H0: Средний пользовательский рейтинг жанра **Action** равен среднему пользовательскому рейтингу жанра **Sports**.

H1: Средний пользовательский рейтинг жанра **Action** **не** равен среднему пользовательскому рейтингу жанра **Sports**.

Альтернативная гипотеза является двусторонней.

Возьмем специальный метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей `scipy.stats.ttest_ind()`.

Параметр `equal_var` укажем как **False**, т.к. есть повод усомниться в равенстве дисперсий.

Параметр `nan_policy` - **omit** даст возможность обойти пропущенные значения.

Критический уровень статистической значимости укажем как **.05**.

```python
alpha = 0.05 # зададим критический уровень статистической значимости

results1 = st.ttest_ind(
    df.loc[df['genre'] == 'Action', 'user_score'], 
    df.loc[df['genre'] == 'Sports', 'user_score'],
    equal_var = False,
    nan_policy = 'omit'
) # проверим гипотезу о равенстве двух независимых совокупностей

print('p-значение:', results1.pvalue)

if results1.pvalue < alpha: # сравним получившееся p-значение с заданным уровнем статистической значимости
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Промежуточный вывод:**

- статистическое исследование показало, что мы не можем опровергнуть нулевую гипотезу, средний пользовательский рейтинг жанра **Action** может не отличаться от среднего пользовательского рейтинга жанра **Sports**.
